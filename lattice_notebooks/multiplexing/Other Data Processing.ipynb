```python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Prepare the data
data = {
    'Frequency': [1600, 1500, 1450, 1430, 1425, 1420, 1412.5, 1405, 1400, 1395, 1387.5, 1375, 1350, 1325, 1300, 1250, 1200, 1100, 1000],
    'S11_Magnitude': [1.593, 1.62, 1.722, 1.735, 1.745, 1.81, 1.807, 1.83, 1.89, 1.855, 1.866, 1.84, 1.772, 1.757, 1.72, 1.732, 1.667, 1.633, 1.55]
}
data_df = pd.DataFrame(data)
data_df['Frequency'] *= 1e6  # Convert MHz to Hz
data_df['S11_Magnitude'] *= -1 * (50/1.2)  # Invert the magnitude

# Lorentzian function definition
def lorentzian(f, f0, gamma, A, offset):
    return A * ((gamma**2) / ((f - f0)**2 + gamma**2)) + offset

# Identify the peak: Now the lowest point (due to inversion)
peak_index = data_df['S11_Magnitude'].idxmin()
fr = data_df.loc[peak_index, 'Frequency']

# Define a window around the peak for fitting
window_size = 50  # Adjust based on dataset granularity
half_window = window_size // 2
start_index = max(peak_index - half_window, 0)
end_index = min(peak_index + half_window, len(data_df) - 1)

freq_window = data_df['Frequency'][start_index:end_index]
s11_mag_window = data_df['S11_Magnitude'][start_index:end_index]

# Initial guesses for Lorentzian parameters
initial_guesses = [fr, 1e6, min(s11_mag_window), np.mean(s11_mag_window)]

# Fit the data to the Lorentzian model
popt, pcov = curve_fit(lorentzian, freq_window, s11_mag_window, p0=initial_guesses)

# Extract fitted parameters
f0_fitted, gamma_fitted, A_fitted, offset_fitted = popt

# Calculate the quality factor
Q_fitted = f0_fitted / (2 * gamma_fitted)

# Plotting the original data and the fitted curve
plt.figure(figsize=(10, 6))
plt.plot(data_df['Frequency'], data_df['S11_Magnitude'], 'o', label='Original Data')
plt.plot(freq_window, lorentzian(freq_window, *popt), 'r--', label='Lorentzian Fit')
plt.axvline(x=f0_fitted, color='green', linestyle='--', label=f'Resonance Frequency = {f0_fitted / 1e6:.3f} MHz')
plt.xlabel('Frequency (Hz)')
plt.ylabel('S11 Magnitude (inverted)')
plt.title('Inverted S11 Magnitude and Lorentzian Fit')
plt.legend()
plt.show()

print(f"Quality factor (Q): {Q_fitted:.2f}")
```
